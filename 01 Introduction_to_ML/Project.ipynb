```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Machine-learning/Video_Games_Sales_as_at_22_Dec_2016.csv')

# Check for missing values in the 'Year_of_Release' column
print(df['Year_of_Release'].isnull().sum())


df['Year_of_Release'].fillna(0, inplace=True)

# df.dropna(subset=['Year_of_Release'], inplace=True)

# Now, safely convert the 'Year_of_Release' column to integers
df['Year_of_Release'] = df['Year_of_Release'].astype(int)

# Check the changes
print(df['Year_of_Release'].head())
```
